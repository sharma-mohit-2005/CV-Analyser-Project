```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Base URL for Internshala job listings
base_url = "https://internshala.com/internships/page-"

# Total number of pages to scrape
total_pages = 237

# Open a file in write mode to save the links
with open("internship_links.txt", "w") as file:
    # Loop over all pages (from page 1 to page 188)
    for page_num in range(1, total_pages + 1):
        # Construct the URL for the current page
        url = f"{base_url}{page_num}/"
        
        # Send a GET request to the website
        response = requests.get(url)

        # Check if the request was successful
        if response.status_code == 200:
            # Parse the HTML content
            soup = BeautifulSoup(response.content, "html.parser")
            
            # Find all job containers (adjust the class if needed)
            job_containers = soup.find_all("div", class_="individual_internship")
            
            # Loop through each job container and extract the "apply" link
            for job in job_containers:
                job_link = job.find("a", class_="job-title-href")  # Find anchor with class 'job-title-href'
                
                if job_link:
                    # Get the href attribute (the job detail link)
                    relative_link = job_link.get('href')
                    
                    # Convert the relative URL to an absolute URL
                    apply_link = urljoin(base_url, relative_link)
                    
                    # Write the absolute link to the file
                    file.write(f"{apply_link}\n")
        else:
            print(f"Failed to retrieve page {page_num}. Status code: {response.status_code}")

print("intership links have been saved to internship_links.txt")
```
